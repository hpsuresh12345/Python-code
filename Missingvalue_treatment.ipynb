```python
import numpy as np
import pandas as pd
```

```python
#create data using dictionary
mydict={"Income":(50,25,15,10,np.NAN,np.NAN,20,10,np.NAN,999),"Age":(19,20,30,np.NAN,np.NAN,25,20,52,46,9999)}
mydict
```

```python
mydata=pd.DataFrame(mydict)
mydata
```

```python
#replace bogus values such as 9999 with NaN
mydata=mydata.replace((9999,999),np.NAN)
mydata
```

```python
# Display values which are 'Null' with 'True'
mydata.isnull()
```

```python
#Drop any rows where there are any columns missing values
new = mydata.dropna(how='any')
new
```

```python
#Drop any rows where there are all columns missing values
new=mydata.dropna(how='all')
new
```

```python
#Replace missing values with mean of respective columns
new=np.round(mydata.fillna(mydata.mean()),1)
new
```

```python
#Replace missing values with mean of respective columns
new=mydata.fillna(mydata.median())
new
```

```python
#Replace missing values with mean of respective columns
new=mydata.fillna({'Age':30,'Income':15})
new
```

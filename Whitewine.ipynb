```python
Y = data.quality
X = data.drop('quality', axis=1)
X.head()
Y.head()
```

# Correlation understanding and Heat map

```python
correlations = data.corr()
print(correlations)
plt.subplots(figsize=(20,20))
sns.heatmap(correlations,cmap='RdYlGn', annot = True)
plt.show();
```

```python
from sklearn.preprocessing import MinMaxScaler
scld=MinMaxScaler(feature_range=(0,1))
arr_scld=scld.fit_transform(X)
data_scld=pd.DataFrame(arr_scld,columns=X.columns)
X.head()
X.describe()
data_scld.head()
data_scld.describe()
data_scld
```

# Creation of Train and test split, splitting Train 75 % and Test 25%

```python
train_x, test_x, train_y,test_y=train_test_split(X,Y, test_size=0.25, random_state=1)
train_x.shape
test_y.shape
train_y.shape
test_x.shape
```

# Creation of Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm
```

```python
lm.fit(train_x,train_y)
```

# Creation of Prediction

```python
predict_test=lm.predict(test_x)
```

```python
print(lm.coef_)
data_m=pd.DataFrame({'X':X.columns,'coeff':lm.coef_})
data_m=data_m.sort_values(by=['coeff'])
data_m
```

```python
data_m.plot(x='X',y='coeff',kind='bar',figsize=(15,10))
plt.show();
```

```python
("RSQUARE Value for TEST data is -")
np.round(lm.score(test_x,test_y)*100,0)
```

# R square values for Train data is 28%

```python
("RSQUARE Value for TRAIN data is -")
np.round(lm.score(train_x,train_y)*100,0)
```

# Mean absolute error calculation which is 0.58

```python
from sklearn.metrics import mean_absolute_error
```

```python
print("Mean Absolute Error(MAE) for TEST data is")
np.round
(mean_absolute_error(test_y, predict_test),0)
```

# Comparison of Predicted and actual - prediction error

```python
fdf=pd.concat([test_x,test_y],1)
fdf['Predicted']=np.round(predict_test,1)
fdf['Prediction_Error']=fdf['quality']-fdf['Predicted']
fdf
```

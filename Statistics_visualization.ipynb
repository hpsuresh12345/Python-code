```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#let's do basic statistics first
iris.describe()
```

```python
np.round(iris.mean(),2)
```

```python
np.round(iris.median(),2)
```

```python
#display non missing counts by each columns
iris.count()
```

```python
#counting frequency for categorical variables
pd.crosstab(index=iris["species"],columns = "Frequency")
```

```python
correlation = iris.corr()
correlation
```

```python
sns.heatmap(correlation, cmap="nipy_spectral",annot=True)
plt.show();
sns.heatmap(correlation, cmap="Blues",annot=True)
plt.show();
```

```python
#compute percentile
np.round(iris.quantile((np.arange(0,1,0.05))),2)
```

```python
#pair plot
sns.pairplot(iris,hue='species',palette="RdYlGn",size=5)
plt.show();
```

```python
#pair plot
sns.pairplot(iris,kind="reg")
plt.show();
```

```python
# Histogram and kernel density estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_width)
plt.show();
```

```python
#box plot
plt.figure(figsize=(15,10))
sns.boxplot(data=iris,orient="v",palette="Set2")
plt.show();
```
